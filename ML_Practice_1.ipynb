```python

def analyze_yes_no_data(df):
    for col in df.columns:
        counts = Counter(df[col])
        print(f"Column: {col}")
        for value, count in counts.items():
            print(f"  {value}: {count} ({count/len(df[col])*100:.2f}%)")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Function to check and visualize the distribution of all attributes in a DataFrame
def check_and_visualize_all_attributes(df):
    num_cols = len(df.columns)
    num_rows = (num_cols + 3) // 4  # Number of rows needed in the subplot grid

    fig, axarr = plt.subplots(num_rows, 4, figsize=(25, 5 * num_rows), squeeze=False)

    for i, column in enumerate(df.columns):
        ax = axarr[i // 4, i % 4]
        if df[column].dtype == 'object' or df[column].nunique() < 10:
            sns.countplot(x=column, data=df, ax=ax)
            ax.set_title(f'{column} Count Distribution')
            ax.set_xlabel('')
            ax.set_ylabel('Count')
        else:
            sns.histplot(df[column], kde=True, color='skyblue', ax=ax)
            ax.set_title(f'{column} Histogram')
            ax.set_xlabel('')
            ax.set_ylabel('Frequency')

    # Turn off axes for any unused subplots
    for j in range(num_cols, axarr.size):
        axarr[j // 4, j % 4].axis('off')

    plt.tight_layout()
    plt.show()

# Call the function with your DataFrame
check_and_visualize_all_attributes(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Function to plot a single histogram showing the progress of each column
def plot_overlay_histograms(df):
    plt.figure(figsize=(12, 8))

    # Loop through each column and plot its histogram
    for column in df.columns:
        if pd.api.types.is_numeric_dtype(df[column]):
            plt.hist(df[column].dropna(), bins=30, alpha=0.5, label=column, edgecolor='black')

    plt.title('Overlay Histograms for Each Column')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.legend()
    plt.grid(True)
    plt.show()

# Call the function with your DataFrame
plot_overlay_histograms(df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Function to create a scatter plot matrix (pairplot) for numeric columns
def plot_scatter_matrix(df):
    # Select only numeric columns
    numeric_df = df.select_dtypes(include=['number'])

    # Create pairplot
    sns.pairplot(numeric_df)
    plt.suptitle('Scatter Plot Matrix for Numeric Columns', y=1.02)
    plt.show()

# Call the function with your DataFrame
plot_scatter_matrix(df)
```

```python
import matplotlib.pyplot as plt

def plot_scatter_dependency(df, productivity_col, other_col):
    if productivity_col in df.columns and other_col in df.columns:
        plt.figure(figsize=(10, 6))
        plt.scatter(df[other_col], df[productivity_col], alpha=0.5, edgecolor='k')
        plt.title(f'{productivity_col} vs {other_col}')
        plt.xlabel(other_col)
        plt.ylabel(productivity_col)
        plt.grid(True)
        plt.show()
    else:
        print(f"One or both columns not found: {productivity_col}, {other_col}")

# Call the function with your DataFrame and specific columns
plot_scatter_dependency(df, 'day feel productive', 'GYM')
plot_scatter_dependency(df, 'day feel productive', 'Date')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Function to plot a box plot for 'Productivity' against a categorical column
def plot_boxplot(df, productivity_col, categorical_col):
    plt.figure(figsize=(12, 6))

    sns.boxplot(x=categorical_col, y=productivity_col, data=df)
    plt.title(f'Box Plot of {productivity_col} by {categorical_col}')
    plt.xlabel(categorical_col)
    plt.ylabel(productivity_col)
    plt.grid(True)
    plt.show()

# Call the function with your DataFrame and specific columns
plot_boxplot(df, 'day feel productive', 'Date')
```

```python
import matplotlib.pyplot as plt

def plot_histogram_date_vs_productivity(df):
    # Ensure 'Date' column is in datetime format
    df['Date'] = pd.to_datetime(df['Date'])
    df['day feel good']=pd.to_numeric(df['day feel productive'])

    # Plotting
    plt.figure(figsize=(12, 6))

    # Plot histogram with 'Date' on x-axis and 'Productivity' on y-axis
    plt.hist(df['Date'], weights=df['day feel good'], bins=30, edgecolor='black', alpha=0.7)

    plt.title('Productivity Distribution Over Time')
    plt.xlabel('Date')
    plt.ylabel('Productivity')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Call the function with your DataFrame
plot_histogram_date_vs_productivity(df)
```
